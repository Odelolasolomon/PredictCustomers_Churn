```python
churn['exited'].value_counts()
```

```python
x=churn.drop(columns=['exited'], axis=1)
y=churn['exited']
```

```python
#standardize the data
Scaler=StandardScaler()

x_std=Scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x_std, y, test_size=0.3, random_state=42)
print(x_train.shape,y_train.shape)
```

### Feature Selection
- Using regularization pattern

```python
#Perform GridsearchCV to tune best-fit LR model
param = {'C': [10**-2, 10**-1, 10**0, 10**1,10**2]}

lr_model= LogisticRegression(penalty='l1', solver='liblinear')

gs_model=GridSearchCV(estimator=lr_model, param_grid=param)

gs_model.fit(x_train, y_train)

#Train a model with best parameters
model=LogisticRegression(**gs_model.best_params_, penalty='l1', solver='liblinear')

model.fit(x_train, y_train)
```

```python
#print the coefficients
coef = model.coef_[0]
coef
```

```python
#remove the redundant features
coef= model.coef_[0]
imp_features = pd.Series(x.columns)[list(coef!=0)]
```

### Modeling

```python
#select proper features now
x=churn.drop(columns=['exited','geography_Spain'], axis=1)
y=churn['exited']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3, random_state=42)
x_train.shape,y_train.shape
```

```python
x_test.shape, y_test.shape
```

```python
#scale the data

x_train=Scaler.fit_transform(x_train)
x_test=Scaler.fit_transform(x_test)
```

```python
#initialize model class

model= LogisticRegression()
model.fit(x_train, y_train)

y_predictions=model.predict(x_test)
y_prob=model.predict_proba(x_test)[:,1]

y_predictions[0:20]
```

```python
#predicted probabilities
y_prob[0:20]
```

```python
plt.boxplot(y_predictions)
plt.show()
```

### Model Evaluation

```python
train_accuracy=model.score(x_train, y_train)
test_accuracy=model.score(x_test, y_test)

print('{} train_accuracy'.format(train_accuracy))
print('{} test_accuracy'.format(test_accuracy))
```

```python
conf_matrix=confusion_matrix(y_test, y_predictions)
print(conf_matrix)
```

```python
class_report=classification_report(y_test, y_predictions)
print(class_report)
```

```python
from sklearn.metrics import roc_curve

#plot tpr,fpr and thresholds
fpr,tpr,thresholds=roc_curve(y_test, y_prob)

plt.plot([0,1], [0,1], 'k--')

plt.plot(fpr, tpr)
plt.xlabel('false positive rate')
plt.ylabel('true positive rate')
plt.title('ROC CURVE')
plt.show()
```

```python
#calculate the area under the curve
from sklearn.metrics import roc_auc_score
area_auc_curve = roc_auc_score(y_test, y_prob)
area_auc_curve
```
